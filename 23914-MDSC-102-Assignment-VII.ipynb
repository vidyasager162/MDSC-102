```python
normalized_df = pd.DataFrame()
```

```python
def Normalize(df):
    for col in df:
        if col == 'Class':
            continue
        series = pd.Series(sp.stats.boxcox(df[col])[0])
        normalized_df[col] = series
```

This code is to estimate the parameter $\mu$'s 95% confidence interval when $\sigma$ is not known. 

```python
def ConfidenceIntervals_mu(df, alpha):
    bounds = {} 
    for col in df:
        row = []
        lb = (df[col].mean() + (sp.stats.t.ppf(alpha/2, df.shape[0] - 1) * (df[col].std() / np.sqrt(df.shape[0]))))
        ub = (df[col].mean() + (sp.stats.t.ppf(1 - (alpha/2), df.shape[0] - 1) * (df[col].std() / np.sqrt(df.shape[0]))))
        row.append(lb)
        row.append(ub)
        bounds[col] = row
    return bounds
```

```python
bounds = ConfidenceIntervals_mu(normalized_df, alpha)
bounds
```

```python
bounds_df = pd.DataFrame(bounds, index=['Lower Bound', 'Upper Bound'])
bounds_df
```

This code is to estimate the parameter $\sigma$'s 95% confidence interval when $\mu$ is not known. 

```python
for col in df:
    print(df[col].var())
```

```python
def ConfidenceIntervals_sigma(df, alpha):
    bounds = {}
    for col in df:
        row = []
        lb = ((df.shape[0]-1)*df[col].var())/sp.stats.chi2.ppf(1 - alpha/2, df.shape[0]-1)
        ub = ((df.shape[0]-1)*df[col].var())/sp.stats.chi2.ppf(alpha/2, df.shape[0]-1)
        row.append(lb)
        row.append(ub)
        bounds[col] = row
    return bounds
```

```python
bounds_sigma = ConfidenceIntervals_sigma(df, 0.05)
bounds_sigma
```

```python
bounds_sigma_df = pd.DataFrame(bounds_sigma, index=['Lower Bound', 'Upper Bound'])
bounds_sigma_df
```
