```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dir(pd)
```

```python
names = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./Resources/wine/wine.data', names=names)
```

```python
df
```

```python
sns.countplot(x="Alcohol", data=df)
```

```python
sns.scatterplot(x="Alcohol", y="Malic acid", data=df, hue="Class")
```

```python
plt.figure(figsize=(20,20))
for index, column in enumerate(df):
    if index > 1:
        plt.subplot(4,3,index-1)
        sns.scatterplot(x="Alcohol", y=column, data=df, hue="Class")
```
