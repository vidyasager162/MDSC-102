```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df = pd.read_csv('./Resources/wine/wine.data', names=names)
```

```python
df
```

```python
sns.boxplot(x="Alcohol", data=df)
```

```python
sns.histplot(x="Alcohol", data=df, kde=True)
```

```python
sns.boxplot(x="Malic acid", data=df)
```

```python
sns.histplot(x="Malic acid", data=df, kde=True)
```

```python
sns.boxplot(x="Class", y="Alcohol", data=df)
```

```python
plt.figure(figsize=(20,20))
for index, column in enumerate(df):
    if index > 0:
        plt.subplot(5,3,index)
        sns.boxplot(x="Class", y=column, data=df)
```
