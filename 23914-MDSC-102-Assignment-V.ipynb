```python
names = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted Wine::::::::::::s","Proline"]
df = pd.read_csv('./Resources/wine/wine.data', names=names)
```

```python
plt.figure(figsize=(20,20))
for index, column in enumerate(df):
    if index > 0:
        plt.subplot(5,3,index)
        sns.boxplot(x="Class", y=column, data=df)
```
