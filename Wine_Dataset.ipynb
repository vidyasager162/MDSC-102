```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
```

```python
names = ["Class","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted Wines","Proline"]
df = pd.read_csv('./Resources/wine/wine.data', names=names)
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
mag_mean = df['Magnesium'].mean()
dil_mean = df['OD280/OD315 of diluted Wines'].mean()
hue_mean = df['Hue'].mean()

mag_median = df['Magnesium'].median()
dil_median = df['OD280/OD315 of diluted Wines'].median()
hue_median = df['Hue'].median()
```

```python
fig, ax = plt.subplots(figsize=(15,20), nrows=3, ncols=2)
fig.suptitle("Plotting Magnesium, OD280/OD315 of diluted Wines and Hue")

sns.boxplot(data=df, x='Magnesium', ax=ax[0,0], color='r')
ax[0,0].set_title('Box Plot for Magnesium')
sns.histplot(data=df, x="Magnesium", kde=True, ax=ax[0,1], color='r')
ax[0,1].axvline(mag_mean, color='black', linestyle='--', label="Mean")
ax[0,1].axvline(mag_median, color='grey', linestyle='-', label="Median")
ax[0,1].legend()
ax[0,1].set_title("Histogram for Magnesium")

sns.boxplot(data=df, x='OD280/OD315 of diluted Wines', ax=ax[1,0], color='g')
ax[1,0].set_title('Box Plot for OD280/OD315 of diluted Wines')
sns.histplot(data=df, x="OD280/OD315 of diluted Wines", kde=True, ax=ax[1,1], color='g')
ax[1,1].axvline(dil_mean, color='black', linestyle='--', label="Mean")
ax[1,1].axvline(dil_median, color='grey', linestyle='-', label="Median")
ax[1,1].legend()
ax[1,1].set_title("Histogram for OD280/OD315 of diluted Wines")

sns.boxplot(data=df, x='Hue', ax=ax[2,0])
ax[2,0].set_title('Box Plot for Hue')
sns.histplot(data=df, x="Hue", kde=True, ax=ax[2,1])
ax[2,1].axvline(hue_mean, color='black', linestyle='--', label="Mean")
ax[2,1].axvline(hue_median, color='grey', linestyle='-', label="Median")
ax[2,1].legend()
ax[2,1].set_title("Histogram for Hue")

plt.savefig('Subplot')
```

```python
df_mag1 = np.sqrt(df['Magnesium'])
df_mag1.skew()
```

```python
df_mag1
```

```python
fig, ax = plt.subplots(figsize=(10,8), nrows=1, ncols=2)
fig.suptitle("Plotting Transformed Magnesium (Square Root)")

sns.boxplot(data=df_mag1, x=df_mag1, ax=ax[0])
ax[0].set_title('Box Plot for Magnesium')
sns.histplot(data=df_mag1, x=df_mag1, kde=True, ax=ax[1])
ax[1].set_title("Histogram for Magnesium")
```

```python
df_mag2 = np.log(df['Magnesium'])
df_mag2.skew()
```

```python
fig, ax = plt.subplots(figsize=(10,8), nrows=1, ncols=2)
fig.suptitle("Plotting Transformed Magnesium (Log)")

sns.boxplot(data=df_mag2, x=df_mag2, ax=ax[0])
ax[0].set_title('Box Plot for Magnesium')
sns.histplot(data=df_mag2, x=df_mag2, kde=True, ax=ax[1])
ax[1].set_title("Histogram for Magnesium")
```
