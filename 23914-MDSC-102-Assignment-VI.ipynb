```python
filtered_df = df.drop(['Class', 'Magnesium'], axis=1)
filtered_df.rename(columns = {'OD280/OD315 of diluted Wines': 'OD280 or OD315 of diluted Wines'}, inplace=True)
filtered_df
```

```python
filtered_df.columns
```

```python
def Plot(df, col):
    fig, ax = plt.subplots(1, 3, figsize=(25,10))
    skew = df[col].skew()
    fig.suptitle(f'Plots for {col}')
    sns.boxplot(data=df[col], x=df[col], ax=ax[0])
    ax[0].set_title('Box Plot')
    sns.histplot(data=df[col], x=df[col], kde=True, ax=ax[1], label=f'Skew: {skew}')
    ax[1].set_title('Histogram')
    ax[1].legend()
    sp.stats.probplot(df[col], plot=ax[2])
    ax[2].set_title('Probability Plot')
    plt.savefig(f'./Wine Dataset Plots/Original Plots/{col}')
```

```python
for col in filtered_df:
    Plot(filtered_df, col)
```

## Analysing the Skewness from the plots
### This analysis is through examining the Box plot, Histogram plot and the Skew Function values.

* Malic Acid is Right Skewed.
* Flavanoids is Right Skewed.
* Nonflavanoid Phenol is Right Skewed.
* Proanthocyanins is Right Skewed.
* Color Intensity is Right Skewed.
* Hue is a little bit Right Skewed.
* OD280 or OD315 of diluted Wines is Left Skewed.
* Proline is Right Skewed.



```python
skewed_df = filtered_df.drop(['Alcohol', 'Ash', 'Alcalinity of ash', 'Total phenols'], axis=1)
skewed_df
```

```python
skewed_df.skew()
```

```python
normalized_df = pd.DataFrame()
```

```python
def Normalize(df):
    for col in df:
        series = pd.Series(sp.stats.boxcox(df[col])[0])
        normalized_df[col] = series
```

```python
Normalize(skewed_df)
normalized_df
```

```python
def Plot_Normalized(skewed_df, normalized_df, col):
    fig, ax = plt.subplots(1, 4, figsize=(25,10))
    original_skew = skewed_df[col].skew()
    normalized_skew = normalized_df[col].skew()
    fig.suptitle(f'Comparing Original vs Normalized Plots for {col}')
    sns.histplot(data=skewed_df[col], x=skewed_df[col], kde=True, ax=ax[0], label=f'Skew: {original_skew}', color= 'r')
    ax[0].set_title('Original Histogram')
    ax[0].legend()
    sns.histplot(data=normalized_df[col], x=normalized_df[col], kde=True, ax=ax[1], label=f'Skew: {normalized_skew}', color='y')
    ax[1].set_title('Normalized Histogram')
    ax[1].legend()
    sp.stats.probplot(skewed_df[col], plot=ax[2], )
    ax[2].set_title('Probability Plot for Original Data')
    sp.stats.probplot(normalized_df[col], plot=ax[3])
    ax[3].set_title('Probability Plot for Normalized Data')
    plt.savefig(f'./Wine Dataset Plots/Normalized Plots/{col}')
    
```

```python
for col in skewed_df:
    Plot_Normalized(skewed_df, normalized_df, col)
```
